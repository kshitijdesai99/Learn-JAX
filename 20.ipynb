# Immutability + Functional Updates + Matrix Operations

This notebook explores JAX's immutability principles and functional programming approach to array operations, including matrix manipulations.

## Why Immutability?

### 🧠 The Core Concept

If arrays could change in-place:
- **JIT compilation** becomes unpredictable
- **Gradients** become messy  
- **Parallelization** breaks

So JAX forces:
- **No in-place mutation**
- **Always return a new array**
- **Functional programming style**

This ensures reproducible computations and enables powerful optimizations.

```python
# Import JAX libraries
import jax  # Main JAX library
import jax.numpy as jnp  # JAX NumPy API (works on CPU/GPU/TPU)
```

### ✅ Exercise 11 — Create a 2D Matrix

Let's start by creating a 2D JAX array to work with throughout these exercises.

```python
data = jnp.array([[10, 1, 24],
                  [20, 15, 14]])
data
```

```python
data.shape
```

### ✅ Exercise 12 — Transpose

Transposing a matrix swaps its rows and columns. This is a fundamental operation in linear algebra.

```python
dataT = data.T
dataT
```

```python
dataT.shape
```

### ✅ Exercise 13 — Indexing

Accessing individual elements in a JAX array works just like in NumPy, using zero-based indexing.

```python
value = data[0, 2]
value
```

### ✅ Exercise 14 — Update a Value

This demonstrates the key difference between NumPy and JAX. NumPy allows in-place modification, while JAX requires creating a new array.

```python
import numpy as np
data_np = np.array(data)
data_np[1,1] = 100 # numpy way
data_np
```

```python
data = data.at[1, 1].set(100)
data
```

### ✅ Exercise 15 — Add to an Element

JAX provides functional methods like `.add()` that return a new array with the modification applied.

```python
data = data.at[0, 0].add(41)
data
```

### ✅ Exercise 16 — Reduce Along Axis

Reduction operations like `min()` can be performed along specific axes to collapse dimensions.

```python
mins = data.min(axis=1)
mins
```

### ✅ Exercise 17 — Select First Row

Slicing operations work similarly to NumPy, allowing you to extract subsets of the data.

```python
data_select = data[0]
data_select
```

### ✅ Exercise 18 — Append Row

Since JAX arrays are immutable, we use functions like `vstack` that return new arrays with the requested modifications.

```python
data = jnp.vstack([data, data_select])
data
```

```python
data
```

## Bonus — Append Column

### Understanding Array Broadcasting

This demonstrates how to add a column to an existing matrix using proper broadcasting techniques.

```python
# Broadcasting: [:, None] converts 1D array (3,) to 2D column vector (3, 1)
# This enables horizontal concatenation with the 2D matrix 'data'
# Final shape: (3 rows, 3+1 columns) = (3, 4)

data_hstack = jnp.hstack([data, data_select[:, None]])
data_hstack
```

### ✅ Exercise 19 — Matrix Multiplication

Matrix multiplication follows the rule: `(m × n) dot (n × p) = (m × p)`. This is a fundamental operation in linear algebra and deep learning.

```python
# (m × n) dot (n × p) = (m × p)
data_prod = jnp.dot(data, dataT)
data_prod
```

### ✅ Exercise 20 — Change Data Type

Sometimes you need to change the data type of arrays for memory efficiency or numerical precision requirements.

```python
data_prod = jnp.array(data_prod, dtype=jnp.float16)
data_prod
```

## Conclusion

### NumPy vs JAX Mindset

**NumPy mindset:**
- Modify arrays in-place
- Mutable data structures

**JAX mindset:**
- Return transformed arrays
- Immutable data structures
- Functional programming approach

This immutability enables JAX's powerful features like JIT compilation, automatic differentiation, and parallel execution.
