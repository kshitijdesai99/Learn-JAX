Now let's import JAX and JAX NumPy. The main difference is that `jax.numpy` operations can be executed on CPU, GPU, or TPU, while regular NumPy only runs on CPU.

## What is a JAX Array? (Exercises 6-8)

JAX arrays are similar to NumPy arrays but with additional capabilities for accelerator execution and automatic differentiation.

### ✅ Exercise 6 — Create a JAX Array

Let's create our first JAX array. Notice it looks similar to a NumPy array, but it's designed to work across different hardware backends.

```python
data = jnp.array([10, 1, 24]) # A NumPy array that can move to accelerators.
data
```

### ✅ Exercise 7 — Check Array Type

JAX arrays have a different internal type than NumPy arrays, even though they behave similarly.

```python
type(data)
```

### ✅ Exercise 8 — Check Array Shape

Like NumPy arrays, JAX arrays have shape information that defines their dimensions.

```python
data.shape
```

### ✅ Exercise 9 — Move Array to Host (CPU Memory)

Sometimes we need to move data from device memory back to host memory, for example to use with regular NumPy or other libraries.

## Host vs Device Memory (Exercises 9-10)

Understanding the difference between host (CPU) memory and device (GPU/TPU) memory is crucial for efficient JAX programming.

```python
data_host = jax.device_get(data)
data_host
```

### ✅ Exercise 10 — Move Data to Device

Let's move the data back to device memory. This is where JAX operations will be executed for optimal performance.

```python
data_device = jax.device_put(data_host)
data_device
```
